```python
import os
from tqdm import tqdm

path_dir = 'G:'
file_list = os.listdir(path_dir)
print(file_list)



for i in tqdm(range(len(file_list))):
    file_name = file_list[i].split(' ')[1]
    file_list1 = file_list[i]
    file_oldname = os.path.join(path_dir, file_list1)
    file_newname = os.path.join(path_dir, file_name)
    os.rename(file_oldname, file_newname)
```
